# Seaborn

Seaborn provides an API on top of Matplotlib that offers sane choices for plot style and color defaults, defines simple high-level functions for common statistical plot types, and integrates with the functionality provided by Pandas DataFrame

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style(style="whitegrid")
```

```python
# Histograms

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

sns.histplot(data=data.x, alpha = 0.4, bins = 'sturges', kde= True)
sns.histplot(data=data.y, alpha = 0.4, bins = 'sturges', kde = True, color='green')
plt.show()
```

```python
sns.kdeplot(data=data.x,fill = True)
sns.kdeplot(data=data.y,fill = True, color= 'green')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data=iris, hue = 'species',size= 'petal_length', alpha = .5, sizes= (4000,100),legend= None, palette= 'Set1')
plt.title('Iris scatter plot', fontsize = 20)
plt.xlabel('sepal_length (cm)')
plt.ylabel('sepal_width (cm)')

plt.show()
```

```python
sns.pairplot(iris, hue='species', size=2.5, palette='Set2')
plt.show()
```

### Scatterplot

```python
plt.figure(figsize=(12,4))
sns.boxplot(x='sepal_length', orient= 'h', data= iris)
plt.title('sepal_length',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(data = iris, x = 'sepal_length', y='species', orient= 'h',
            palette= 'Set2')
plt.title('Altura',fontsize = 18)
plt.xlabel('Metros')
plt.show()
```
