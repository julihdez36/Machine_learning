# Visualization with Matplotlib

Matplotlib is a multi-platform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack. It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting.


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Setting Styles
# We'll use the plt.style direcive to choose aesthetic styles for pur figures

plt.style.available
```

```python
plt.style.use('ggplot')

x = np.linspace(1,10,100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show() # looks for all currently active figure objects and opens a interactive window to display it.
```

## Two interfaces for building graphics

In Matplotlib we can find two ways or interfaces to build graphs: a convenient MATLAB-style state-based interface, and a more powerful object-oriented interface (VanderPlas, 2016)

```python
# Matplotlib was originally written as a Python alternative for MATLAB users
# Matlab-style interface

plt.figure() # create a plot figure

# Create the first of two panels and set current axis

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set the current axis

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
# Object-oriented interface

## If you want more control over your figure

# First create a grid of plots
# ax will be array of two axes objects

fig, ax = plt.subplots(2)

# Call plot() method on the appropiate object
ax[0].plot(x, np.sin(x), color = 'blue')
ax[1].plot(x, np.cos(x), color = 'purple')

plt.show()
```

### Line plots

Let's start with a simple linear functionL $y = f(x)$.

```python
# For all Matplotlib plots, we start by creating a figured an axes

fig = plt.figure() # It's a simple container that contains all the objects 
ax = plt.axes() # It's a bounding box with ticks and labels
```

```python
# Now we can to plot some data on the blonding box

fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
plt.show()
```

```python
# Also we can graph with the pylab interface

plt.plot(x,np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

## Adjusting the plot: colors, styles, axes limits

```python
# Adjusting the plot: line colors and styles

plt.plot(x, x + 4, linestyle='-', color = 'purple')  # solid
plt.plot(x, x + 5, linestyle='--', color = 'red') # dashed
plt.plot(x, x + 6, linestyle='-.', color = 'yellow') # dashdot
plt.plot(x, x + 7, linestyle=':', color = 'blue')  # dotted

plt.show()
```

```python
# Axes limits

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# Labeling plots and legend

plt.title("Sine and cosine Curve")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.plot(x, np.sin(x),  linestyle= '--', label='sin(x)')
plt.plot(x, np.cos(x), linestyle = ':',label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
# Object-oriented interface to plotting

ax = plt.axes()
ax.plot(x,np.sin(x),linestyle = ':', color = 'b')
ax.set(xlim=(0,10),ylim=(-2,2),xlabel= 'x',ylabel='f(x)',
       title = 'A sin function') # method to set all these properties at once
plt.show()
```

### Scatter plots

  Instead of points being joined by line segments, here the points are represented individually with a dot, circle, or other shape.

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), 'o', color='purple')
plt.show()
```

```python
np.random.seed(123)

x = np.random.randn(100)
y = np.random.randn(100)

fig, ax = plt.subplots()
scatter = ax.scatter(x,y,c = y, s = x*1000, alpha= .3, cmap='viridis')
ax.set(title='Scatter plot', xlim= (0,3))
fig.colorbar(scatter, ax = ax) # Show color scale
plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data


plt.scatter(data[:,0],data[:,1], alpha=.2,
           s=100*data[:,3],c=iris.target,cmap='viridis')
plt.colorbar()
```

### Visualizing Errors

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.plot(x,y, color = 'red')
plt.errorbar(x, y, yerr=dy, fmt='.k', ecolor= 'lightgray', elinewidth=3)
plt.plot()
```
