### Leyes de la suma de matrices y del producto poe escalares

Sean $A,B \in \mathcal{M}_{mxn}(\mathbb{K})$ La suma de matrices cumple las siguientes propiedades:
1. Asociativa: $(A+B)+C = A+(B+C)$
2. Conmutativa: $A+B = B+A$
3. Elemento neutro: $A+0_{mxn} = A$
4. Elemento opuesto: $A+(-A) = 0_{mxn}$

Y el producto por escalares cumplen:
5. Distributiva respecto a la suma de matrices: $\alpha (A+B) = \alpha A + \alpha B$
6. Distributiva respecto de la suma de escalares: $(\alpha + \beta)A = \alpha A + \beta A$
7. Asociativa con respecto del producto por escalares: $(\alpha \beta) A = \alpha (\beta A)$
8. Unidad del cuerpo, $1 \in \mathbb{K}$, cumple que $1A = A$
